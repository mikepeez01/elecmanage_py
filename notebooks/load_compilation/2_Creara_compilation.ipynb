```python
df_alias = pd.read_excel(path_alias_elec)
```

```python
data = Creara.get_inventory(category=Creara.CategoryInventory.SUPPLY)
```

```python
from utils.utils_var import insert_alias

df_id = pd.DataFrame(data)
df_id
df_id['alias'] = df_id.apply(lambda row: insert_alias(row['cups'], df_alias)[1], axis=1)

df_id['cliente'] = df_id.apply(lambda row: insert_alias(row['cups'], df_alias)[0], axis=1)
df_id['tarifa'] = df_id['tarifa'].apply(lambda row: str(row)[:3] + ' ' + str(row)[-2:])
df_id.set_index('id', inplace=True)
# df_id = df_id[~(df_id['alias'] == 'WARNER BQ')]
# df_id = df_id[df_id['cliente'] == 'Altan Pharma']
(df_id) = df_id[(df_id['alias'] != 'WARNER DW')]
(df_id)
```

```python
df_load = pd.DataFrame()
for index, row in df_id.iterrows():

    id = index
    print(row['cups'])
    # df['cups'] = df_id.loc[id, 'cups']
    df_hour = Creara.get_metering(id=id,
                        start_date=start_date,
                        end_date=end_date,
                        data_soruce=str(Creara.DataSource.COUNTER),
                        resolution=str(Creara.Resolution.HOURLY))
    if not df_hour.empty:
        df_hour['resolution'] = str(Creara.LoadColumnResolution.HOURLY)
        df_hour['datetime'] = df_hour['datetime'] - timedelta(hours=1)
    df_cuarter = Creara.get_metering(id=id,
                        start_date=start_date,
                        end_date=end_date,
                        data_soruce=str(Creara.DataSource.COUNTER),
                        resolution=str(Creara.Resolution.QUARTER_HOURLY))
    if not df_cuarter.empty:
        df_cuarter['resolution'] = str(Creara.LoadColumnResolution.QUARTER_HOURLY)
        df_cuarter['datetime'] = df_cuarter['datetime'] - timedelta(minutes=15)
    # filtramos los cups de Warner según el día que se instalaron los contadores de Crera
    if (id == 118485) or (id == 195403):
        df_cuarter = df_cuarter[df_cuarter['datetime'] >= datetime(2025, 6, 1)]
        df_hour = df_hour[df_hour['datetime'] >= datetime(2025, 6, 1)]

    df = pd.concat([df_cuarter, df_hour], ignore_index=True)
    if not df.empty:
        df['cups'] = row['cups']

    df_load = pd.concat([df_load, df], ignore_index=True)

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
df_load.to_pickle(f"{path_load_elec_clean}/load_creara{timestamp}.pkl")
 
```
